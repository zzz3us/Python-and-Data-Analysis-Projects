### Dropping unnecessary columns

Let's now explore the data to find out what else do we need to clean up.

```python
autos.describe(include = "all")
```

As we can see, several columns contain irrelevant data and can easily be dropped: "seller" and "offer_type" contain only one value repeated over all the rows ("privat" and "Angebot" respectively).

```python
autos = autos.drop(["seller","offer_type"], axis = 1)
```

```python
autos.loc[:,"nr_of_pictures"].isnull
```

```python
autos.loc[:,"nr_of_pictures"].isna().mean()
```

Additionally, all the values in the "nr_of_pictures" seems to be "0.0" so the column will also be dropped.

```python
autos = autos.drop(["nr_of_pictures"], axis = 1)
```

### Converting string values to numeric values

Next, two columns: "price" and "odometer" store numeric values as text and must be further cleaned, by removing any non-numeric characters, converting the datatype of the column to numeric and changing the columns' names to reflect the changes:

```python
autos['price'].head()
```

Removing non-numeric parts ("$" and ",") of the string and changing the dtype of the column to "int":

```python
autos["price"] = (autos["price"].str.replace("$","").str.replace(",","").astype(int))

autos['price'].head()
```

```python
autos["odometer"].head()
```

Removing non-numeric parts ("km", ",") of the string and changing the dtype to "int":

```python
autos["odometer"] = (autos["odometer"].str.replace("km","").str.replace(",","").astype(int))

autos["odometer"].head()
```

Renaming the column to "odometer_km" to reflect the changes:

```python
autos.rename({"odometer":"odometer_km"}, axis = 1, inplace = True)
```

```python
autos.head()
```

### Further look into "odometer" and "price" columns

Before moving forward, let's take a quick look at the "odometer_km" and "price" columns again:

```python
autos["odometer_km"].value_counts()
```

Rounded value grouped into even intervals suggest that the sellers had to choose from pre-set options.

```python
autos["price"].value_counts()
```

```python
autos["price"].unique().shape
```

```python
autos["price"].describe()
```

In the "price" column we find 1421 items with a price of 0. Even though eBay is an auction platform it is highly unlikely for users to post their offers starting at 0 price point. This can either mean that there were no price value available for these records or that the data was somehow lost in the process of aggregating it.

```python
autos["price"].value_counts().sort_index(ascending=False).head(20)
```

There is also a number of records with an unusually high prices. While prices ranging into hundred of thousands euros are possible, anything over 1 million seems highly unlikely. Also, given that the number increase steadily up to 350000 euros then jump sharply into 999990 euros might suggest that everything above 350k euros is unreliable data and as such should be removed.
